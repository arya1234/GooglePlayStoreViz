```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./kaggle/googleplaystore.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.drop(['Current Ver','Android Ver'],inplace=True,axis=1)
```

```python
df
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
df['Type'] = df['Type'].fillna('Free')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

# DataFrame Category Visualization

```python
plt.figure(figsize=[12,6])
sns.countplot(df['Category'],order=df['Category'].value_counts().index[0:14])
plt.xticks(fontsize=12, rotation=90)
plt.show()
```

# We see that the Top 3 Category in Apps are 'Family','Game' and 'Tools'

```python
df_beta = df.sort_values(by='Reviews',ascending=False)
```

```python
df['Reviews'] = df['Reviews'].astype(float)
```

# Count plot for Free and Paid Apps 

```python
sns.countplot(x='Type',data=df)
plt.show()
```

# Maximum of the Apps are Free as compared to Paid

# We are taking Paid apps into other Dataframe called beta


```python
beta = df[df['Type'] == 'Paid']
```

```python
beta
```

```python
beta['Price'] = beta['Price'].str.replace('$','')
beta['Price'] = beta['Price'].astype(float)
```

# Category Visualization 

```python
plt.figure(figsize=[15,9])
sns.countplot(beta['Category'],order=beta['Category'].value_counts().index[:10])
plt.xticks(rotation=90)
plt.show()
```

# In the 'Paid' apps, majority of the Apps are of 'Family' Category and the 2nd Highest are of 'Medical' Category

```python
beta.info()
```

```python
beta = beta.sort_values(by='Price',ascending=False)
```
